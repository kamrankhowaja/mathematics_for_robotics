### PMF (Probability Mass Function)
- let $X$ be a discrete random variable with range $R_X=\{ x_1, x_2 ,..... x_n\}$ (i.e finite or infinite). The function

$P_X(X_k) = P(X=X_k) for k =1,2,3 .....$ is called the Probability Mass Function (PMF) of X.

- it gives you the probability distribution of random variable X (Discrete)
##### Properties of PMF:
- $O \leq P_X(x) \leq 1$
- $\sum P_X(X) = 1$
- if you need find a probability of individual then find it via $\sum P_X(X) = 1$

### CDF (Cummulative Distribution Function)
- The cumulative distribution function (CDF) of a random variable is another method to describe the distribution of random variables.
- The advantage of the CDF is that it can be defined for any kind of random variable (discrete, continuous, and mixed).

$F_{x} = P(X \leq x) = \sum_{v_{n} \leq x} P_n$

##### Properties of CDF:
- $f(-\infty)=0, f(\infty)=1$
- CDF is a not decreasing Function 
- for any $X_k \epsilon R_X $ its given as $ F_x(X_k)-F_X(X_k- \epsilon) = f(X=x_k)$ . that is always small
- if $P(a < x \leq b) = P(X \leq b) - P(x \leq a) = F_X(b)-F_X(a)$

#### Example 
**I toss a coin twice.**

Let $X$ be the number of observed heads. Find the CDF of $X$.

set of possible outcomes = $\{ HH,HT,TH,TT \}$ , Assign Random Variables considering that you find head
- For (X = 0) = $\{TT\}$ , No heads with probability $P(X=0) = \frac{1}{4}$
- For (X = 1) = $\{HT , TH\}$ , one head with probability $P(X=1) = \frac{1}{4} + \frac{1}{4}= \frac{2}{4} = \frac{1}{2}$
- For (X = 2) = $\{HH\}$ , for two heads with probability $P(X=2) = \frac{1}{4} $

According to CDF it is given as $F_{x} = P(X \leq x) = \sum_{v_{n} \leq x} P_n$

- For $x<0$, it cannot be negative so $P(X \leq x) = 0 $ , as $F_x(x)=0$
- For $0 \le x<1$ , this valid for our first situation $(X \leq 0)$ , so $F_x(x) = P(X=0)= \frac{1}{4}$
- For $1 \le x<2$ , this valid for our first and second situation $(X \leq 1)$ , so $F_x(x) = P(X=0) + P(X=1)= \frac{1}{4} + \frac{1}{2} = \frac{3}{4}$
- For $x \geq 2$ , this valid for all sitautaions $(X \geq 2)$ , so $F_x(x) = P(X=0) + P(X=1) + P(X=2)= \frac{1}{4} + \frac{1}{2} + \frac{1}{4} = 1$

##### Final Expression

$$ 
F_X(x) =
\begin{cases} 
0, & x < 0 \\
\frac{1}{4}, & 0 \leq x < 1 \\
\frac{3}{4}, & 1 \leq x < 2 \\
1, & x \geq 2
\end{cases}
 $$

#### Expected Value of a Function of a Random Variable

$E[g(X)] = \sum_{x_k \epsilon R_x} g(x_k).P_X(x_k)$

$$
R_X = \left\{ 0, \frac{\pi}{4}, \frac{\pi}{2}, \frac{3\pi}{4}, \pi \right\}

$$
