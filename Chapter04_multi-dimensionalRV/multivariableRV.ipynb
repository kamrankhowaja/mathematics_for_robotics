### Multivariate PMF/PDF and CDF

##### Probability Mass Function

The PMF or PDF of a bivariate random variable gives the probability that the two random variables each take a certain value.

The PMF of a bivariate random variable is a function that gives the probability that the components of $X=x$ takes the value $X_1=x_1$ and $X_2=x_2$. it is given as,

$$f_{x_1,x_2}(x_1,x_2) = P(X=x_1).P(X=x_2)$$

Two Properties of PMF(Probability Mass Function):
- $f_{x_1,x_2}(x_1,x_2) \geq 0$
- $\sum_{x_1} \sum_{x_2} f_{x_1,x_2}(x_1,x_2) = 1 $



##### Commulative Distribution Function

The CDF of bivariate random variable returns the total probability that each component is less then or equal to given value. it is given as 
$$f_{x_1 x_2}(x_1,x_2) = P(X_1 < x_1 , X_2 < x_2)$$

#### Marginal Distribution :
The marginal distribution gives the distribution of a single variable in a joint distribution. In the case of bivariate distribution. it is calculated by the sum of all probabilities for $X_1$ across all the values in the support of $X_2$. The Marginal distribution is given as : 
$$f_{X_1}(x_1) = \sum_{x_2 \epsilon R(X_2)}  f_{X_1 X_2}(x_1,x_2)$$ , $$f_{X_2}(x_2) = \sum_{x_1 \epsilon R(X_1)}  f_{X_1 X_2}(x_1,x_2)$$

**Marginalization:** Reduction of joint distribution to a lower dimensional one via a summation process is called marginalization.

##### Calculating (Marginal Distribution)

For $X_1$:
$$ P(X_1 = -10) = 5\% + 10\% + 20\% = 35\%$$
$$ P(X_1 = 0) = 5\% + 10\% + 5\% = 20\%$$
$$ P(X_1 = 10) = 30\% + 15\% + 0\% = 45\%$$

For $X_2$:
$$ P(X_2 = 1) = 5\% + 5\% + 30\% = 40\%$$
$$ P(X_2 = 0) = 10\% + 10\% + 15\% = 35\%$$
$$ P(X_2 = -1) = 20\% + 5\% + 0\% = 25\%$$

**Does this shows Joint PMF??????????** NO! as $5\% \not= 35\% * 40\%$

It is clear that the two variables are not independent because multiplying their marginal PMFs does not lead us back to the joint PMF.

#### Conditional distribution:

The conditional distributions describe the probability of an outcome of a random variable conditioned on the other random variable taking a particular value. the relation bond return/ positive analyst.

$$f_{X_1|X_2}(x_1|X_2 = x_2) = P(X_1 = x_1 | X_2 = x_2) = (-10\% . \frac{5\%}{40\%}) + (0\% . \frac{5\%}{40\%}) + 10\% . \frac{30\%}{40\%} = 6.25$$

#### Covariance 
Relation between two Random variables can be calculated by:
$$ cov[X_1,X_2] = E[((X_1)- E[X_1]) ((X_2)- E[X_2])]$$
$$ cov[X_1,X_2] = E[X_1 X_2] - E[X_1]E[X_2]$$

#### Co-relation 
ratio of the covariance b/w two random variables and the product of their standard deviations
$$ coorelation(X_1,X_2) = \frac{cov[X_1,X_2]}{std.deviation X_1.std.deviation X_2}$$

#### Linear Transformation
It does not change the coorelation b/w random variables it just scale it with constants 
$$ cov(a+bx_1,c+dx_1) = b.d.cov(X_1,X_2)$$
